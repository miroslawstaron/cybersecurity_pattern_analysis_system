```python
# import the libraries and read the csv file from database/embeddings_codebert.csv into a dataframe
import pandas as pd
import numpy as np

df = pd.read_csv('../database/embeddings_codebert_vanilla.csv', sep='$')

df.head()
```

```python
# add one more column with the name of the vulnerability
df['vulnerability'] = 'input_validation'
```

```python
# now read the csv file ../database/embeddings_codebert_sql_injection.csv into a dataframe
# add the vulnerability column with 'sql_injection' as value
# and append the dataframe to the previous one
df2 = pd.read_csv('../database/embeddings_codebert_sql_injection.csv', sep='$')
df2['vulnerability'] = 'sql_injection'
df = df.append(df2)
```

```python
# return this as json df.groupby('vulnerability').count()['filename']
print(df.groupby('vulnerability').count()['filename'].to_json())

```

```python
# save the dataframe
df.to_csv('../database/embeddings_codebert.csv', sep='$', index=False)
```
